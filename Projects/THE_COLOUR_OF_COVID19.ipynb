```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# group the data by State and sum the total cases
State_cases = CRT.groupby('State')['Cases_Total'].mean()
#State_cases = CRT.groupby('State')['Cases_Total'].fillna(0).max()
# create a new dataframe with the state and total cases
State_cases_CRT = pd.DataFrame({'State': State_cases.index, 'Cases_Total': State_cases.values})

# sort the dataframe by total cases in descending order
State_cases_CRT = State_cases_CRT.sort_values('Cases_Total', ascending=False)

# create a heatmap using seaborn
plt.figure(figsize=(19,13))
sns.heatmap(State_cases_CRT.set_index('State'), cmap='YlOrRd', annot=True, fmt='g')
plt.title('Total COVID-19 Cases by State', fontsize=16)
plt.xlabel('Cases_Total', fontsize=16)
plt.ylabel('State', fontsize=16)
plt.show() # shows that CA, TX, FL, NY, IL have the maximiun cases 
#.max means The chart suggests that some states, such as California, Texas, Florida, and New York, have significantly higher numbers of COVID-19 cases compared to other states.
#.mean would give you a sense of the average number of cases per state.
```

```python
# group the data by State and sum the total cases
State_cases = CRT.groupby('State')['Deaths_Total'].mean()

# create a new dataframe with the state and total cases
State_cases_CRT = pd.DataFrame({'State': State_cases.index, 'Deaths_Total': State_cases.values})

# sort the dataframe by total cases in descending order
State_cases_CRT = State_cases_CRT.sort_values('Deaths_Total', ascending=False)

# create a heatmap using seaborn
plt.figure(figsize=(13,12))
sns.heatmap(State_cases_CRT.set_index('State'), cmap='YlOrRd', annot=True, fmt='g')
plt.title('Total COVID-19 Cases by State', fontsize=16)
plt.xlabel('Deaths_Total', fontsize=16)
plt.ylabel('State', fontsize=16)
plt.show() #NY has the highest average death total 
#It shows that some states have significantly higher numbers of COVID-19 deaths than others, such as New York, California, and Texas, while some states have lower numbers of deaths, such as Alaska, Hawaii, and Vermont. 
#The heatmap provides a visual representation of the severity of the pandemic in different states and can be useful for identifying areas that may require more resources and attention to address the impact of the pandemic.
```

```python
# Calculate the total number of cases - sum does not work need to find the mean or average
total_cases = CRT['Cases_Total'].mean(skipna=True)

# Calculate the number of cases for each race/ethnicity
White_cases = CRT.loc[CRT['Cases_White'].notna(), 'Cases_Total'].mean()
Black_cases = CRT.loc[CRT['Cases_Black'].notna(), 'Cases_Total'].mean()
Latinx_cases = CRT.loc[CRT['Cases_Latinx'].notna(), 'Cases_Total'].mean()
Asian_cases = CRT.loc[CRT['Cases_Asian'].notna(), 'Cases_Total'].mean()
AIAN_cases = CRT.loc[CRT['Cases_AIAN'].notna(), 'Cases_Total'].mean()
NHPI_cases = CRT.loc[CRT['Cases_NHPI'].notna(), 'Cases_Total'].mean()
Multiracial_cases = CRT.loc[CRT['Cases_Multiracial'].notna(), 'Cases_Total'].mean()
Other_cases = CRT.loc[CRT['Cases_Other'].notna(), 'Cases_Total'].mean()

# Calculate the number of cases for each additional column
Unknown_cases = CRT.loc[CRT['Cases_Unknown'].notna(), 'Cases_Total'].mean()
#Hispanic_cases = CRT.loc[CRT['Cases_Ethnicity_Hispanic'].notna(), 'Cases_Total'].sum()
#NonHispanic_cases = CRT.loc[CRT['Cases_Ethnicity_NonHispanic'].notna(), 'Cases_Total'].sum()
#EthnicityUnknown_cases = CRT.loc[CRT['Cases_Ethnicity_Unknown'].notna(), 'Cases_Total'].sum()

# Calculate the proportion of each race/ethnicity
prop_white_cases = White_cases / total_cases
prop_black_cases = Black_cases / total_cases
prop_latinx_cases = Latinx_cases / total_cases
prop_asian_cases = Asian_cases / total_cases
prop_AIAN_cases = AIAN_cases / total_cases
prop_NHPI_cases = NHPI_cases / total_cases
prop_Multiracial_cases = Multiracial_cases / total_cases
prop_Other_cases = Other_cases / total_cases

# Calculate the proportion of each additional column
prop_Unknown_cases = Unknown_cases / total_cases
#prop_Hispanic_cases = Hispanic_cases / total_cases
#prop_NonHispanic_cases = NonHispanic_cases / total_cases
#prop_EthnicityUnknown_cases = EthnicityUnknown_cases / total_cases

# Create a DataFrame with the proportions
prop_df = pd.DataFrame({
    'case_type': ['White', 'Black', 'Latinx', 'Asian', 'AIAN', 'NHPI', 'Multiracial', 'Other', 'Unknown'], #'Hispanic', 'Non-Hispanic', 'Ethnicity Unknown'],
    'proportion': [prop_white_cases, prop_black_cases, prop_latinx_cases, prop_asian_cases, prop_AIAN_cases, prop_NHPI_cases, prop_Multiracial_cases, prop_Other_cases, prop_Unknown_cases]  
    #prop_Hispanic_cases, prop_NonHispanic_cases, prop_EthnicityUnknown_cases]
}) 
 
# Plot the stacked bar chart
plt.figure(figsize=(13, 12))
sns.set_style("whitegrid")
sns.set_palette("Paired")
sns.barplot(x="proportion", y="case_type", data=prop_df, orient='h', palette='Paired')
plt.title('Proportions of COVID-19 Cases by Race/Ethnicity')
plt.xlabel('Proportion')
plt.ylabel('Race/Ethnicity')
plt.show()
#The chart highlights the disparities in the proportion of COVID-19 cases among different racial and ethnic groups, with Black and Latinx individuals being disproportionately affected compared to White individuals.
```

```python
# Calculate the total number of deaths
total_deaths = CRT['Deaths_Total'].mean(skipna=True)

# Calculate the number of deaths for each race/ethnicity
White_deaths = CRT.loc[CRT['Deaths_White'].notna(), 'Deaths_Total'].mean()
Black_deaths = CRT.loc[CRT['Deaths_Black'].notna(), 'Deaths_Total'].mean()
Latinx_deaths = CRT.loc[CRT['Deaths_Latinx'].notna(), 'Deaths_Total'].mean()
Asian_deaths = CRT.loc[CRT['Deaths_Asian'].notna(), 'Deaths_Total'].mean()
AIAN_deaths = CRT.loc[CRT['Deaths_AIAN'].notna(), 'Deaths_Total'].mean()
NHPI_deaths = CRT.loc[CRT['Deaths_NHPI'].notna(), 'Deaths_Total'].mean()
Multiracial_deaths = CRT.loc[CRT['Deaths_Multiracial'].notna(), 'Deaths_Total'].mean()
Other_deaths = CRT.loc[CRT['Deaths_Other'].notna(), 'Deaths_Total'].mean()

# Calculate the number of deaths for each additional column
Unknown_deaths = CRT.loc[CRT['Deaths_Unknown'].notna(), 'Deaths_Total'].mean()
#Hispanic_deaths = CRT.loc[CRT['Deaths_Ethnicity_Hispanic'].notna(), 'Deaths_Total'].sum()
#NonHispanic_deaths = CRT.loc[CRT['Deaths_Ethnicity_NonHispanic'].notna(), 'Deaths_Total'].sum()
#EthnicityUnknown_deaths = CRT.loc[CRT['Deaths_Ethnicity_Unknown'].notna(), 'Deaths_Total'].sum()

# Calculate the proportion of each race/ethnicity
prop_white_deaths = White_deaths / total_deaths
prop_black_deaths = Black_deaths / total_deaths
prop_latinx_deaths = Latinx_deaths / total_deaths
prop_asian_deaths = Asian_deaths / total_deaths
prop_AIAN_deaths = AIAN_deaths / total_deaths
prop_NHPI_deaths = NHPI_deaths / total_deaths
prop_Multiracial_deaths = Multiracial_deaths / total_deaths
prop_Other_deaths = Other_deaths / total_deaths

# Calculate the proportion of each additional column
prop_Unknown_deaths = Unknown_deaths / total_deaths
#prop_Hispanic_deaths = Hispanic_deaths / total_deaths
#prop_NonHispanic_deaths = NonHispanic_deaths / total_deaths
#prop_EthnicityUnknown_deaths = EthnicityUnknown_deaths / total_deaths

# Create a DataFrame with the proportions
prop_df = pd.DataFrame({
    'death_type': ['White', 'Black', 'Latinx', 'Asian', 'AIAN', 'NHPI', 'Multiracial', 'Other', 'Unknown'], #'Hispanic', 'Non-Hispanic', 'Ethnicity Unknown'],
    'proportion': [prop_white_deaths, prop_black_deaths, prop_latinx_deaths, prop_asian_deaths, prop_AIAN_deaths, prop_NHPI_deaths, prop_Multiracial_deaths, prop_Other_deaths, prop_Unknown_deaths] #prop_Hispanic_deaths, prop_NonHispanic_deaths, prop_EthnicityUnknown_deaths]
})
#prop_df['proportion'] = prop_df['proportion'].astype(float)
# Plot the stacked bar chart
plt.figure(figsize=(13, 12))
sns.set_style("whitegrid")
sns.set_palette("Paired")
sns.barplot(x="proportion", y="death_type", data=prop_df, orient='h', palette='Paired')
plt.xlabel('Proportion')
plt.ylabel('Race/Ethnicity')
plt.title('Proportions of COVID-19 Deaths by Race/Ethnicity') 
#The chart shows that White and Black individuals have the highest proportions of COVID-19 deaths, followed by Latinx individuals. 
#The chart also shows that AIAN, NHPI, Multiracial, and Other groups have smaller proportions of deaths compared to the larger racial/ethnic groups. 
#The Unknown category is also included in the chart, which represents individuals whose race/ethnicity was not reported or is unknown. 
#The chart can help to identify disparities in COVID-19 deaths among different racial/ethnic groups and inform public health interventions to address these disparities.

#The chart shows the proportion of deaths for each race/ethnicity relative to the total number of deaths. The stacked bar chart allows for easy comparison between the different groups 
#and shows that there are disparities in COVID-19 deaths by race/ethnicity. The chart shows that Black, Latinx, and AIAN communities have a higher proportion of COVID-19 deaths compared to 
#their representation in the total population. 
```

```python
# calculate the proportion of cases for each race/ethnicity
total_cases = CRT[['Cases_White', 'Cases_Black', 'Cases_Latinx', 'Cases_Asian', 'Cases_AIAN', 'Cases_NHPI', 'Cases_Multiracial','Cases_Other', 'Cases_Unknown',]].max(skipna=True)
prop_cases = total_cases / CRT['Cases_Total'].max(skipna=True)
#prop_cases = total_cases/total_cases

# create a DataFrame with the proportion of cases for each race/ethnicity
prop_cases_df = pd.DataFrame(prop_cases).reset_index()
prop_cases_df.columns = ['Race/Ethnicity', 'Proportion of Cases']

# create the stacked bar chart using Seaborn
fig, ax = plt.subplots(figsize=(13, 12))
sns.barplot(x='Proportion of Cases', y='Race/Ethnicity', data=prop_cases_df, color='purple')
ax.set_xlabel('Proportion of Cases')
ax.set_ylabel('Race/Ethnicity')
ax.set_title('COVID Proportion of Cases by Race/Ethnicity') #The chart highlights that Black and Latinx individuals have a higher proportion of COVID-19 cases compared to their representation in the population, 
#while White and Asian individuals have a lower proportion of cases. The chart suggests that COVID-19 is disproportionately affecting Black and Latinx communities.
```

```python
# calculate the proportion of deaths for each race/ethnicity
total_deaths = CRT[['Deaths_White', 'Deaths_Black', 'Deaths_Latinx', 'Deaths_Asian', 'Deaths_AIAN', 'Deaths_NHPI', 'Deaths_Multiracial','Deaths_Other', 'Deaths_Unknown',]].max()
prop_deaths = total_deaths / CRT['Deaths_Total'].max(skipna=True)

# create a DataFrame with the proportion of deaths for each race/ethnicity
prop_deaths_df = pd.DataFrame(prop_deaths).reset_index()
prop_deaths_df.columns = ['Race/Ethnicity', 'Proportion of Deaths']

# create the stacked bar chart using Seaborn
fig, ax = plt.subplots(figsize=(13, 12))
sns.barplot(x='Proportion of Deaths', y='Race/Ethnicity', data=prop_deaths_df, color='purple')
ax.set_xlabel('Proportion of Deaths')
ax.set_ylabel('Race/Ethnicity')
ax.set_title('COVID Proportion of Deaths by Race/Ethnicity') ##The chart highlights the disproportionate impact of COVID-19 on different racial and ethnic groups, with certain groups experiencing higher rates of mortality than others.
```
